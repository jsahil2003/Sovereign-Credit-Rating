```python
# importing all required libraries
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
# reading datasets
```

```python
factors = pd.read_excel('/Users/sahiljadhav/Documents/scr_proj/data.xlsx') 
gs = pd.read_excel('/Users/sahiljadhav/Documents/scr_proj/global_south.xlsx')
g20 = pd.read_excel('/Users/sahiljadhav/Documents/scr_proj/g20.xlsx')
```

```python
# adding necessary binary columns 
```

```python
factors['in_g20'] = factors['countries'].isin(g20['g20']).astype(int)
factors['in_global_south'] = factors['countries'].isin(gs['global_south']).astype(int)
```

```python
# defining the main dataset df
```

```python
df = factors
```

```python
# creating the dataset required for regression as df_reg
```

```python
columns_to_drop = [
    'countries', 'moodys_rating', 'fitch_rating', 's_rating',
    'moody_numeric', 'fitch_numeric', 's_p_numeric',
    'basu_rating', 'mean_rating', 'gdp_cur'
]
df_reg = df.drop(columns = columns_to_drop)
```

```python
# transforming gdp to ln(gdp) as values are too large 
```

```python
df_reg['ln_gdp'] = np.log(df['gdp_cur'])
```

```python
# creating the target variable
```

```python
dep_var = df['moody_numeric'] 
```

```python
df_reg['target'] = dep_var
```

```python
# spliting the data
```

```python
train_data , test_data = train_test_split(df_reg , test_size = 0.20 , random_state = 134)
```

```python
x_train = train_data.drop('target' , axis =1)
y_train = train_data['target']
```

```python
# creating a multiple linear regression model
```

```python
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
# testing the model
```

```python
x_test = test_data.drop('target' , axis = 1)
y_test = test_data['target']
pred = model.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_test,pred)
```
